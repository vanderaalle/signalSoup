```python
import numpy as np

def calcsfnratio(band):
    sf = []
    for  u in [a.get_sf() for a in band.agents]:
        for w in u:
            sf.append(w)
    return len(sf)/len(band.agents)




s = []
for j in tqdm(range(10, 110, 10)):
    r = []
    for x in range(100):
        b = Band(j)
        k = []
        sf = []
        for i in range(10000):
            b.turn()
            k.append(calcsfnratio(b))
        r.append(k)
    r1 = np.array(r)
    r1 = sum(r1)/len(r)
    s.append(r1)
```
